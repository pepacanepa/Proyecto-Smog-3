```python
# 📊 Gráfico de barras dobles (lado a lado) con Altair
import pandas as pd
import altair as alt

# Cargar el archivo CSV
df = pd.read_csv("basedatos3.csv", sep=";")

# Reemplazar guiones por 0 en La Pintana
df = df.replace("-", "0")

# Convertir valores de PM2,5 a numéricos
df["PM2,5 primer año"] = pd.to_numeric(df["PM2,5 primer año"].astype(str).str.replace(",", "."), errors="coerce")
df["PM2,5 último ingreso"] = pd.to_numeric(df["PM2,5 último ingreso"].astype(str).str.replace(",", "."), errors="coerce")

# Reestructurar datos en formato largo
df_long = df.melt(
    id_vars=["Estación ", "Comuna"],
    value_vars=["PM2,5 primer año", "PM2,5 último ingreso"],
    var_name="Momento",
    value_name="PM25"
)

# Renombrar valores para la leyenda
df_long["Momento"] = df_long["Momento"].replace({
    "PM2,5 primer año": "Primer ingreso",
    "PM2,5 último ingreso": "Último ingreso"
})

# Crear gráfico de barras lado a lado
chart = alt.Chart(df_long).mark_bar(size=20).encode(
    x=alt.X("Estación :N", title="Estación de monitoreo", sort=None),
    y=alt.Y("PM25:Q", title="Niveles de PM2,5", scale=alt.Scale(domain=[0, 51])),
    color=alt.Color("Momento:N", title="Momento de medición",
                    scale=alt.Scale(domain=["Primer ingreso", "Último ingreso"],
                                   range=["#4C78A8", "#F78AC2"])),  # azul y rosado
    xOffset="Momento:N",  # 🔹 barras lado a lado
    tooltip=["Estación ", "Comuna", "Momento", "PM25"]
).properties(
    width=800,
    height=500,
    title="Comparación de niveles de PM2,5 por estación (primer vs último ingreso)"
).configure_axis(
    labelAngle=-45
)

chart.interactive()
```

```python
# 📊 Script en Google Colab con Altair
import pandas as pd
import altair as alt

# Cargar el archivo CSV
df = pd.read_csv("basedatos3.csv", sep=";")

# Normalizar columnas de fechas y valores
df["primer ingreso"] = pd.to_datetime(df["primer ingreso"], errors="coerce", dayfirst=True)
df["último ingreso"] = pd.to_datetime(df["último ingreso"], errors="coerce", dayfirst=True)

# Convertir valores de PM2,5 a numéricos (reemplazando comas por puntos)
df["PM2,5 primer año"] = pd.to_numeric(df["PM2,5 primer año"].astype(str).str.replace(",", "."), errors="coerce")
df["PM2,5 último ingreso"] = pd.to_numeric(df["PM2,5 último ingreso"].astype(str).str.replace(",", "."), errors="coerce")

# Reestructurar datos en formato largo
df_long = pd.DataFrame({
    "Estación": pd.concat([df["Estación "], df["Estación "]], ignore_index=True),
    "Comuna": pd.concat([df["Comuna"], df["Comuna"]], ignore_index=True),
    "Fecha": pd.concat([df["primer ingreso"], df["último ingreso"]], ignore_index=True),
    "PM25": pd.concat([df["PM2,5 primer año"], df["PM2,5 último ingreso"]], ignore_index=True),
    "Tipo": ["Primer ingreso"] * len(df) + ["Último ingreso"] * len(df)
})

# Crear gráfico de líneas
chart = alt.Chart(df_long).mark_line(point=True).encode(
    x=alt.X("Fecha:T", title="Tiempo", scale=alt.Scale(domain=["2000-01-01", "2026-05-05"])),
    y=alt.Y("PM25:Q", title="Niveles de PM2,5", scale=alt.Scale(domain=[0, 51])),
    color=alt.Color("Estación:N", title="Estación de monitoreo"),
    tooltip=["Estación", "Comuna", "Fecha", "PM25"]
).properties(
    width=800,
    height=500,
    title="Evolución de PM2,5 en estaciones de monitoreo SINCA"
)

chart.interactive()
```
